```python
# feature selection
movies_data['content']= movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['title']+' '+movies_data['cast']+' '+movies_data['director']
movies_data.head()
```

```python
movies_data.isnull().sum()
```

```python
movies_data['content'].fillna(' ',inplace=True)
```

```python
movies_data.isnull().sum()
```

**(i)Stemming using stopwords**

```python
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
import nltk
nltk.download('stopwords')
```

```python
# make corpus after removing symbols and numbers, tokenization , lower all alphabets and lemmatization 

from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()

corpus=[]
for i in range(0,len(movies_data)):
    review=re.sub('[^a-zA-Z]',' ',movies_data['content'][i]) # make review sentence with the charecters having a-z or A-Z latter  
    review=review.lower() # make every charecter lower case
    review=review.split()
    review=[lemmatizer.lemmatize(word) for word in review if word not in set(stopwords.words('english'))]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

**(ii)TF-IDF model to get sentence to vector**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv=TfidfVectorizer(ngram_range=(1,3),
                  max_features=4)
```

```python
X=cv.fit_transform(corpus)
cv.vocabulary_ 
```

```python
corpus[0]
```

```python
X[0].toarray()
```

```python
X=X.toarray()
X
```

# Generate Similarity Index

```python
#using Cosine Similarity
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(X)
print(similarity)
```

```python
similarity.shape
```

### Making recommandation using steps 

```python
# taking input which movie user want to watch
wanted_movie = input('Movie you want : ')
```

```python
# list of movie names 
l1=list(movies_data['title'])
```

```python
# choose some closest possible movies name as per the user given input
import difflib    # this library will try to detect something similarities if the movie spelling is wrong or the movie name does not exist
closest_possible_movies = difflib.get_close_matches(wanted_movie,l1)
closest_possible_movies
```

```python
# closest movies name as per the user given input 
closest_movie = closest_possible_movies[0]
closest_movie
```

```python
# index of the closest movie
index= movies_data[movies_data.title == closest_movie]['index'].values[0]  #values[0] will try to store the index value into the similarity score list 
index
```

```python
# calculate the similarity score of other movies with my closest movies
similarity_score = list(enumerate(similarity[index]))
similarity_score
```

```python
#sort similar movies according to similarity score
similar_movies = sorted(similarity_score,key = lambda x:x[1], reverse = True) 
similar_movies
```

```python
print('0','.',movies_data.title[index])
for i in range(1,20):
    index = similar_movies[i][0]
    print(i,'.',movies_data.title[index])
```

# Movie Recommendation system

```python
wanted_movie = input('Movie you want : ') 
l1=list(movies_data['title'])
import difflib
closest_possible_movies = difflib.get_close_matches(wanted_movie,l1)
closest_movie = closest_possible_movies[0]
index= movies_data[movies_data.title == closest_movie]['index'].values[0]
similarity_score = list(enumerate(similarity[index]))
similar_movies = sorted(similarity_score,key = lambda x:x[1], reverse = True)

print('0','.',movies_data.title[index])
for i in range(1,20):
    index = similar_movies[i][0]
    print(i,'.',movies_data.title[index])
```
