```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns     
```

```python
movies_data = pd.read_csv('movies.csv')
movies_data
```

```python
movies_data.info()
```

# Data Preprocessing

```python
# feature selection
movies_data['content']= movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['title']+' '+movies_data['cast']+' '+movies_data['director']
movies_data['content']
```

```python
new_data=movies_data['content']
new_data
```

```python
#Handling missing value
new_data.isnull().sum()
```

```python
new_data.fillna(' ',inplace=True)
new_data.isnull().sum()
```

```python
# Feature extruction
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
import nltk
nltk.download('stopwords')
# printing the stopwords in English
print(stopwords.words('english'))
```

```python
# stemming the data
port_stem = PorterStemmer()
def stemming(new_data):
    stemmed_data = re.sub('[^a-zA-Z]',' ',new_data)
    stemmed_data = stemmed_data.lower()
    stemmed_data = stemmed_data.split()
    stemmed_data = [port_stem.stem(word) for word in stemmed_data 
                       if not word in stopwords.words('english')]
    stemmed_data = ' '.join(stemmed_data)
    return stemmed_data
```

```python
new_data= new_data.apply(stemming)
print(new_data)
```

```python
#Tf-Idf
vectorizer = TfidfVectorizer()
vectorizer.fit(new_data)
new_data = vectorizer.transform(new_data)
print(new_data)
```

# Generate Similarity Index

```python
#using Cosine Similarity
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(new_data)
print(similarity)
```

```python
similarity.shape
```

### Making recommandation using steps 

```python
wanted_movie = input('Movie you want : ')
```

```python
l1=list(movies_data['title'])
l1
```

```python
# choose some closest possible movies name as per the user given input 
import difflib     # this library will try to detect something similarities if the movie spelling is wrong or the movie name does not exist 
closest_possible_movies = difflib.get_close_matches(wanted_movie,l1)
closest_possible_movies
```

```python
closest_movie = closest_possible_movies[0]
closest_movie
```

```python
index= movies_data[movies_data.title == closest_movie]['index'].values[0]  #values[0] will try to store the index value into the similarity score list 
index
```

```python
# calculate the similarity score of other movies with my closest movies
similarity_score = list(enumerate(similarity[index]))
similarity_score
```

```python
#sort similar movies according to similarity score
similar_movies = sorted(similarity_score,key = lambda x:x[1], reverse = True) 
similar_movies
```

```python
for i in range(0,20):
    index = similar_movies[i][0]
    print(i,'.',movies_data.title[index])
```

# Movie Recommendation system

```python
wanted_movie = input('Movie you want : ') 
l1=list(movies_data['title'])
import difflib
closest_possible_movies = difflib.get_close_matches(wanted_movie,l1)
closest_movie = closest_possible_movies[0]
index= movies_data[movies_data.title == closest_movie]['index'].values[0]
similarity_score = list(enumerate(similarity[index]))
similar_movies = sorted(similarity_score,key = lambda x:x[1], reverse = True) 
for i in range(0,20):
    index = similar_movies[i][0]
    print(i,'.',movies_data.title[index])
    
```
